```python
from sympy import *
import inspect
f,g=Function("f",real=True),Function("g",real=True)
x=Symbol("x")

print(inspect.getsource(Function))
```

```python
class Lp(Function):
    arg=(1,2)
    @classmethod
    def eval(cls, p):
        if p.is_Number:
            if p.is_zero:
                return oo
            elif p is S.Infinity:
                return maximum(f(x),x)
            elif p >0:
                return integrate(f(x),(x,-oo,oo))
            else:
                return

    def _eval_is_real(self):
        return self.args[0].is_real

p=2
Lp(f,2)
```
